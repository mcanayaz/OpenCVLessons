```python
#Filtreler

#2D Convolution

import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('images/opencv_logo.png')
kernel = np.ones((5,5),np.float32)/25
dst = cv2.filter2D(img,-1,kernel) #-1 verildiğinde çıkış imgesinin giriş imgesiyle aynı derinliğe sahip olması sağlanır.

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Ortalama Filtresi')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#Bulanıklaştırma
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('images/opencv_logo.png')

blur = cv2.blur(img,(5,5))
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Bulanık')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#Gauss Filtresi


import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('images/opencv_logo.png')

blur = cv2.GaussianBlur(img,(5,5),0)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Gauss Filtresi')
plt.xticks([]), plt.yticks([])
plt.show()

 
#Not: # 1 boyutlu Gaussian kernel oluşturma   a = cv2.getGaussianKernel(7, 1)
```

```python
#Median Filtresi


import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('images/noisy2.png')

median = cv2.medianBlur(img,5)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(median),plt.title('Median Filtresi')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python



import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('images/noisy2.png')

blur = cv2.bilateralFilter(img,9,75,75)
#ikinci parametre-Her piksel komşusunun çapı.
#üçüncü parametre- Renk uzayındaki değeri. Değer ne kadar büyük olursa birbirine uzak renkler karışmaya başlar.
#dördüncü paramtre-Koordinat uzayındaki değeri  . Değeri ne kadar büyük olursa, renklerinin sigmaColor aralığında olması koşuluyla, daha fazla piksel birlikte karışacaktır.
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Bilateral Filtresi')
plt.xticks([]), plt.yticks([])
plt.show()

```
